Functions to extract relevent data from Jain data file

```python
def data_extract_Jain(data_file):
    df = pd.read_csv(data_file)
    df.drop([0, 4])
    df.rename(columns={'VL': 'Light'}, inplace=True)
    df.rename(columns={'VH': 'Heavy'}, inplace=True)
    df.rename(columns={"Fab Tm by DSF (°C)": 'Temp'}, inplace=True)

    light_seq = df['Light'].values.tolist()
    heavy_seq = df['Heavy'].values.tolist()
    temp = df['Temp'].values.tolist()

    return light_seq, heavy_seq, temp


def seq2BERTy(heavy,light):
    
    embed_list = []
    length = len(light)

    for i in range(0,length):
  
        sequences = {
        "H": heavy[i], 
        "L": light[i]
      }

        emb = igfold.embed(
        sequences=sequences,
      )
        berty = emb.bert_embs
        encoded = torch.sum(berty,dim=1)
    
        embed_list.append(encoded)
        final = torch.cat(embed_list, dim=0)
    
    return final
```
